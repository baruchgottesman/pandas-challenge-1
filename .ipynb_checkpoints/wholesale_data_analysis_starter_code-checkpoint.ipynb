```python
# Store the client ids of those top 5 clients in a list.
# for this, we'll do the same thing - by valuecount, by nlargest - but this time we'll use the "tolist" function

top_5_clients = df['client_id'].value_counts().nlargest(5).index.tolist()
top_5_clients
```

## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
# first step is to create within the dataframe a "subtotal" column, and define it as unit * quantity
df['subtotal'] = df['unit_price'] * df['qty']
# asks to just print "head" (top five) of the three identified columns
df[['unit_price', 'qty', 'subtotal']].head()
```

```python
# Create a column for shipping price.
# same deal - create a new "total weight" so we can work on the second step of deciding on the
# shipping price.
df['total_weight'] = df['unit_weight'] * df['qty']

# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
# Now that we have a poundage we can create a shipping price and segregate it by whether it is more or less than 50 lbs
#  also made sure to start with numpy at the start of this project, so went back and added it with import numpy as np
df['shipping_price'] = np.where(df['total_weight'] > 50,
                                df['total_weight'] * 7,
                                df['total_weight'] * 10)
df[['qty', 'unit_weight', 'total_weight', 'shipping_price']].head(10)

```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
#  we start with a new "total price" column, which will be the total price (subtotal from two cells above; plus shipping - 
#  with tax added
df['total_price'] = ((df['subtotal'] + df['shipping_price']) * 1.0925).round(2)
df[['subtotal', 'shipping_price', 'total_price']].head(10)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
#  basically the same thing as before but instead of what's charged to the client (price x qty + shipping)
df['line_item_costs'] = ((df['unit_price'] * df['qty']) + df['shipping_price'])
df[['unit_price', 'qty', 'shipping_price', 'line_item_costs']].head(10)
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = ((df['total_price'] - df['line_item_costs']))
df[['line_item_costs', 'total_price', 'profit']].head(10)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
# Before we used the following function - df['client_id'].value_counts().nlargest(5)
# So the first thing we'll do is use research to find a way to convert that into a list
def get_top_clients(): 
    return df['client_id'].value_counts().nlargest(5).index

top_client_ids = list(get_top_clients())
# and we'll test that by printing the "list" called top_client_ids
print(top_client_ids)

#  now we'll use the same function we did above with the test orders, except instead of just one item, it'll organize all their spending
# and the total quantity of what they are buying

for client_id in top_client_ids: 
    top_clients_df = df[df['client_id'] == client_id]
    total_quantity_for_top_clients = top_clients_df['qty'].sum()
    total_spending_for_top_clients = top_clients_df['total_price'].sum().round(2)

    print(f"Client ID: {client_id}")
    print(f"Sum of Quantity Order through all orders: {total_quantity_for_top_clients}")
    print(f"Total Amount spent through all order: ${total_spending_for_top_clients:,.2f}")
                                  
                                
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
# we're going to borrow the original data frames, including the listing of top 5 clients in the "top_clients" issue in the prior cell
#  and now create a new "summary" dataframe
Summary_data_of_top_clients = []
for client_id in top_client_ids:
    client_df = df[df['client_id'] == client_id]
    # now we'll just create a bunch of different columns for "units purchased", total shipping price, ttl revenue and ttl profit
    # by defining each of those new columns
    total_units = client_df['qty'].sum()
    total_shipping = client_df['shipping_price'].sum().round(2)
    total_revenue = client_df['total_price'].sum().round(2)
    total_cost = (client_df['unit_cost'] * client_df['qty']).sum()
    total_profit = (total_revenue - total_cost - total_shipping).round(2)
    
    Summary_data_of_top_clients.append({
        'client_id': client_id,
        'total_units': total_units,
        'total_shipping': total_shipping,
        'total_revenue': total_revenue,
        'total_profit': total_profit
    })
summary_df = pd.DataFrame(Summary_data_of_top_clients)
summary_df.set_index('client_id', inplace=True)
print(summary_df)
    
```

```python
# Format the data and rename the columns to names suitable for presentation.
#  we can do it column by column (using the information from the prior cell)
# but instead we'll use an automatic command that will capitalize and remove _ from each column
#  we'll have to add the millions separately

df.columns = df.columns.str.replace('_', ' ').str.title()
df.columns


# Define the money columns. 
# In this we'll define "money_columns"

money_columns = ['Unit Price', 'Unit Cost', 'Subtotal', 'Shipping Price', 'Total Price', 'Line Item Costs', 'Profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    return f'${value / 1e6:.2f}M'

# Apply the currency_format_millions function to only the money columns. 
for col in money_columns:
    if col in df.columns:
        df[col] = df[col].apply(currency_format_millions)
    else:
        print(f"Column '{col}' not found in DataFrame")


# Rename the columns to reflect the change in the money format.
df = df.rename(columns={col: col + ' (Millions)' for col in money_columns if col in df.columns})


df.head(15)
```
